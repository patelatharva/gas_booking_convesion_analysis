```python
X, y = df.drop(["Conversion", "EventTimestamp", "SearchId", "DealerId", "deviceid", "GeoPoint", "lat", "long", "LocationTwoDecimals", "LocationThreeDecimals", "DealerOpened", "RecommendationType"], axis=1), df["Conversion"]
```

```python
X.shape
```

```python
df.select_dtypes(['object']).columns
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1)
```

```python
def prepare_inputs(X_train, X_test):
    cat_vars = list(X_train.select_dtypes(include=['object']).copy().columns)
    num_vars = list(X_train.select_dtypes(exclude=['object']).copy().columns)
    oe = OrdinalEncoder(cols=cat_vars)
    oe.fit(X_train)
    X_train_enc = oe.transform(X_train)
    X_test_enc = oe.transform(X_test)
    return X_train_enc, X_test_enc
```

```python
def prepare_targets(y_train, y_test):
    le = LabelEncoder()
    le.fit(y_train)
    y_train_enc = le.transform(y_train)
    y_test_enc = le.transform(y_test)
    return y_train_enc, y_test_enc
```

```python
def select_features(X_train, y_train, X_test, k="all"):
    fs = SelectKBest(score_func=chi2, k=k)
    fs.fit(X_train, y_train)
    X_train_fs = fs.transform(X_train)
    X_test_fs = fs.transform(X_test)
    return X_train_fs, X_test_fs, fs
```

```python
# prepare input data
X_train_enc, X_test_enc = prepare_inputs(X_train, X_test)
```

```python
# prepare output data
y_train_enc, y_test_enc = prepare_targets(y_train, y_test)
```

```python
# feature selection
X_train_fs, X_test_fs, fs = select_features(X_train_enc, y_train_enc, X_test_enc)
```

```python
for i in range(len(fs.scores_)):
    print(list(X_train_enc.columns)[i],str(fs.scores_[i]))
```

```python
feature_selection_df = pd.DataFrame(data={"feature": X_train.columns, "score": fs.scores_}).sort_values(by=["score"], ascending=[False])
fig, ax = plt.subplots(figsize=(15,5))
plt.bar(feature_selection_df["feature"], feature_selection_df["score"])
plt.xticks(list(feature_selection_df["feature"]), rotation=90)
plt.show()
```

```python
X_train_final, X_test_final, fs = select_features(X_train_enc, y_train_enc, X_test_enc, k = 6)
```
